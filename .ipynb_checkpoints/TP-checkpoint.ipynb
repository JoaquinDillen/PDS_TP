### Procesamento Digital de Sinal

```python
import numpy as np
from scipy.signal import lfilter, lfilter_zi, filtfilt, butter, freqz, welch, spectrogram
from scipy.io.wavfile import read
from matplotlib import pyplot as plt
```

```python
filename = 'audio/rec3.wav'
fs, audio1 = read(filename)
f, t, S1 = spectrogram(audio1, fs, window='flattop', nperseg=fs//10, noverlap=fs//20, scaling='spectrum', mode='magnitude')
print('filename: ', filename)
print('Data length (s): ', t[-1])
print('Sampling frequency (samples/s): ', fs)

plt.rcParams['figure.figsize'] = 16,10
plt.pcolormesh(t, f, S1)
plt.xlabel('time(s)')
plt.ylabel('frequency(Hz)')
plt.show()
```

```python
wn = 1000 / (fs/2)
b3,a3 = butter(4, wn)
audio1_filt = lfilter(b3, a3, audio1)
```

```python
f, t, S1 = spectrogram(audio1_filt, fs, window='flattop', nperseg=fs//10, noverlap=fs//20, scaling='density', mode='magnitude')
print('filename: ', filename)
print('Data length (s): ', t[-1])
print('Sampling frequency (samples/s): ', fs)

plt.rcParams['figure.figsize'] = 16,10
# pyplot.pcolormesh(t, f[:450], S1[:450][:])
plt.pcolormesh(t, f, S1)
plt.xlabel('time(s)')
plt.ylabel('frequency(Hz)')
plt.show()
```
